```python
from matplotlib import pyplot as plt
from matplotlib import style

# use some style in your graph
style.use('ggplot')
x1 = [1,7,22,65,66]
y1 = [23,45,67,98,120]
x2 = [2,6,9,20,32,40]
y2 = [34,67,90,120,110,60]
plt.plot(x1,y1,label="first")
plt.plot(x2,y2,label="second")
plt.legend()
plt.title("2 Curve Depiction - by AstroDestroyer")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```
